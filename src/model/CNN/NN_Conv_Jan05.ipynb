```python
import pandas
import numpy as np

# import data
file_name = "training_0.0.1.csv"
# load the dataset
dataframe = pandas.read_csv(file_name,header=None)
dataset = dataframe.values
print(dataset)
print("################")
num_row = dataset.shape[0]
num_col = dataset.shape[1]
X = dataset[1:,0:num_col-2]
Y = dataset[1:,num_col-1]
print(X)
print("################")
print(Y)
print(num_row,num_col)

X = np.expand_dims(X,axis=0)
Y = np.expand_dims(Y,axis=0)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv1D,Flatten,MaxPooling1D, Dropout
EPOCHS = 100
BATCH_SIZE = 40

model = Sequential()

model.add(Conv1D(64,kernel_size=2,activation='relu',input_shape=(num_row-1,num_col-2)))
model.add(Conv1D(32,kernel_size=2,activation='relu'))
model.add(Dropout(3))
model.add(MaxPooling1D(pool_size=3))
model.add(Flatten())
model.add(Dense(50,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(129,activation='softmax'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X,Y,epochs=EPOCHS,batch_size=BATCH_SIZE)
```
